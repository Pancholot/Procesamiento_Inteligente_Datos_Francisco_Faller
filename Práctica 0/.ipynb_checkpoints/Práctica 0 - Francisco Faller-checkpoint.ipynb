```python
lista = [1,2,3]
contador = 0
for i in lista:
    print(i)
print("Aquí termina la primera lista for")
for i in range(0,11):
    print(i)
print("Aquí termina la segunda lista for")
for i in range(10, -1, -1):
    print(i)
print("Aquí termina la tercera lista for")
while contador <= 10:
    print(contador) 
    contador += 1 
print("Aquí termina la lista con while")
```

```python
def cuadrado(n):
    return n*n
def factorial(n):
    if n <= 1:
        return n
    else:
        return n * factorial(n-1)
print("El cuadrado de 5 es:", cuadrado(5))
print("El factorial de 5 es:", factorial(5))
```
