```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("hotel_data.csv")
data.shape
```

```python
data.head()
```

```python
X_train = np.array(data["Habitaciones_Ocupadas"]).reshape(-1, 1)
y_train = np.array(data["Ingreso_Restaurante"])

r = linear_model.LinearRegression()
r.fit(X_train, y_train)
y_pred = r.predict(X_train)

print("Coeficiente de correlaci칩n: ", r.coef_)
print("Interseccion: ", r.intercept_)
print("Error cuadr치tico de la media: ", mean_squared_error(y_train, y_pred))
print("Varianza: ", r2_score(y_train, y_pred))
```

```python
x = X_train.ravel()
plt.figure(figsize=(8, 5))
plt.scatter(x, y_train, color="steelblue", s=80, label="Datos")
plt.plot(x, y_pred, color="red", linewidth=3, label="Pendiente")
plt.xlabel("Habitaciones Ocupadas")
plt.ylabel("Ingreso del Restaurante")
plt.title("Correlaci칩n entre Habitaciones Ocupadas e Ingreso del Restaurante")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
