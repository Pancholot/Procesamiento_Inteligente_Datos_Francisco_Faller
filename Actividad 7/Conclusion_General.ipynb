## Conclusión y reflexión final

Con base en los resultados obtenidos en los tres modelos implementados durante esta práctica, considero que el modelo más útil para un responsable de políticas públicas es la **Red Bayesiana**.

En un principio la regresión lineal nos permite identificar relaciones entre variables y medir el efecto promedio de ciertos factores sobre las ventas, lo cual es útil para situaciones en las que se quieran análisis económicos generales. La regresión logística, por su parte, nos permitió clasificar empresas como exitosas o no exitosas con una buena capacidad predictiva, resultando en que pudiesemos identificar perfiles de riesgo o probabilidad de éxito.

Sin embargo, como había pensando desde un principio, la Red Bayesiana no solo mide asociaciones, sino que modela explícitamente las relaciones causales que existen entre las variables de nuestro sistema o base de datos, como lo fue las que utilizamos. Esto es especialmente relevante en el ámbito de políticas públicas, ya que nos permite simular escenarios y analizar cómo cambiaría la probabilidad de éxito si se implementaran programas de financiamiento o capacitación.

En general, al hablar de un responsable de políticas públicas, según lo investigado, necesitan tener la capacidad de evaluar distintos escenarios y comprender los efectos directos e indirectos de posibles intervenciones antes de implementarlas. Por ello, aunque los tres modelos aportan información valiosa, la Red Bayesiana resulta más útil para la toma de decisiones estratégicas orientadas al diseño de programas de apoyo empresarial.
