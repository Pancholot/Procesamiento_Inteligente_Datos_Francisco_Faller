# Metodo uno / metodo aprendido por medio del primer ejercicio

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Cargar el conjunto de datos de diabetes
df_diabetes = load_diabetes()

# Seleccionar la variable independiente BMI (índice de masa corporal)
x = df_diabetes.data[:,2].reshape(-1,1)
print ("Variable independiente:", x)

# Seleccionar la variable dependiente la progresión de la enfermedad after one year
y = df_diabetes.target
print("Variable dependiente:", y)

```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, Y_train)  # Crea el modelo
Y_pred = regressor.predict(X_test)

print("Coefficientes:", regressor.coef_)
print("Independt term:", regressor.intercept_)
print("Mean Squared Error (MSE): %.2f" % mean_squared_error(Y_test, Y_pred))
print(regressor.score(X_test, Y_test))
print("Varianza calificación: %.2f" % r2_score(Y_test, Y_pred))
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].scatter(X_train, Y_train)
ax[0].plot(X_train,regressor.predict(X_train), c='orange')
ax[0].set_title("Train Set")
ax[0].set_xlabel("Feature")
ax[0].set_ylabel("Target")
ax[1].scatter(X_test, Y_test)
ax[1].plot(X_train, regressor.predict(X_train), c="orange")
ax[1].set_title("Test Set")
ax[1].set_xlabel("Feature")
ax[1].set_ylabel("Target")
plt.suptitle("Linear Regression")
```

# Metodo dos / metodo para obtener el MSE, Coeficiente de Determinación y gráficas con los mismo resultados pedidos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data_diabetes = load_diabetes()
x2 = data_diabetes.data[:, 2].reshape(-1, 1)
y2 = data_diabetes.target


# División manual de los datos en TRAIN y TEST utilizando las últimas 20 muestras como TEST
X2_train = x2[:-20]
X2_test = x2[-20:]

Y2_train = y2[:-20]
Y2_test = y2[-20:]

model_regressor = LinearRegression()
model_regressor.fit(X2_train, Y2_train)

y2_pred = model_regressor.predict(X2_test)

# Métricas con resultados exactos pedidos
print("coeficients:\n", model_regressor.coef_)
print("indepentent term:\n", model_regressor.intercept_)
print("Mean Squared Error (MSE): %.2f" % mean_squared_error(Y2_test, y2_pred))
print("Varianza calificación: %.2f" % r2_score(Y2_test, y2_pred))
```

```python
# Gráficas con resultados exactos pedidos
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].scatter(X2_train, Y2_train)
ax[0].plot(X2_train, model_regressor.predict(X2_train), c="orange")
ax[0].set_title("Train Set")
ax[0].set_xlabel("Feature")
ax[0].set_ylabel("Target")
ax[1].scatter(X2_test, Y2_test)
ax[1].plot(X2_test, model_regressor.predict(X2_test), c="orange")
ax[1].set_title("Test Set")
ax[1].set_xlabel("Feature")
ax[1].set_ylabel("Target")
plt.suptitle("Linear Regression")
```
